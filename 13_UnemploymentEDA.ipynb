```python
!pip install pandas
!pip install numpy
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('/content/output.csv')

# Display the first few rows
df.head()
```

```python
# Display the basic information about the dataset
print("\nBasic information about the dataset:")
print(df.info())

# Summary statistics
print("\nSummary statistics of the dataset:")
print(df.describe())

# Checking for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Data types of the columns
print("\nData types of the columns:")
print(df.dtypes)
```

```python
# Distribution of unemployment rates
plt.figure(figsize=(10, 6))
sns.histplot(df['Rate'], bins=30, kde=True)
plt.title('Distribution of Unemployment Rates')
plt.xlabel('Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 counties with the highest unemployment rates
top_10_unemployment = df.nlargest(10, 'Rate')
print("\nTop 10 counties with the highest unemployment rates:")
print(top_10_unemployment[['County', 'State', 'Rate']])
```

```python
# Unemployment rate by state
state_avg_unemployment = df.groupby('State')['Rate'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(x=state_avg_unemployment.values, y=state_avg_unemployment.index, palette='viridis')
plt.title('Average Unemployment Rate by State')
plt.xlabel('Average Unemployment Rate')
plt.ylabel('State')
plt.show()
```

```python
# Scatter plot of Year vs. Unemployment Rate
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Year', y='Rate', data=df, alpha=0.6)

# Adding title and labels
plt.title('Scatter Plot of Year vs. Unemployment Rate')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')

# Display the plot
plt.show()
```
