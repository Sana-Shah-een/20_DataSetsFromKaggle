```python
# Import the dataset
data1=pd.read_csv("/content/googleplaystore.csv")
data1
```

```python
data2=pd.read_csv("/content/googleplaystore_user_reviews.csv")
data2
```

```python
data=data1.merge(data2,how='inner',on='App')
data
```

```python
# #Display the first few rows of the dataset
data.head()
```

```python
# Get basic information about the dataset
data.info()
```

```python
# Summary statistics of the dataset
data.describe()
```

```python
# Check for missing values
data.isnull().sum()
```

**Which is the most used Android version?**

```python
# Find the most installed category
most_installed_category = data.groupby('Android Ver')['Installs'].sum().idxmax()
most_installed_category
```

** Which all categories of mob apps are existing from this dataset?**

```python
# Get unique categories
categories = data['Category'].unique()
categories
```

**Which is the most installed category of mob app from this dataset?**

```python
# Find the most installed category
most_installed_category = data.groupby('Category')['Installs'].sum().idxmax()
most_installed_category
```

**How many total categories are there?**

```python
data['Category'].nunique()
```

**Which apps have got 5-star ratings?**

```python
data.groupby('Rating')['App'].value_counts(ascending=False)
```

**Which are the top 10 mob apps based on ratings?**

```python
data.sort_values(by='Rating',ascending=False).head(10)
```
