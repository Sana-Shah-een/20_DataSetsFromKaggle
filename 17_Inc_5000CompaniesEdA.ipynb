```python
!pip install pandas
!pip install numpy
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Import the dataset file
data=pd.read_csv("/content/INC 5000 Companies 2019.csv")
data
```

```python
# Displays first few rows of the dataset
data.head()
```

```python
# Displays information about the columns of the dataset
data.info()
```

```python
data['revenue']= data["revenue"].replace({'Million':"","Billion":"*1000"},regex=True).map(pd.eval)
print(data["revenue"].dtype)
```

```python
# Displays the numeric data of the dataset
data.describe()
```

```python
# Checks for the missing data
data.isnull().sum()
```

**Data AnalysisÂ¶**

**What's the average revenue among companies on the list? Broken down by industry?**

```python
avg_revenue=data.groupby('industry')['revenue'].mean().reset_index()
avg_revenue
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 8))
avg_revenue.plot(kind='bar')
plt.title('Averege revenue of industries')
plt.xlabel('Industry')
plt.ylabel('Average revenue')
plt.xticks(rotation=90)
plt.show()
```

```python
ind_counts=data['industry'].value_counts()
ind_counts
```

```python
# Identify the most and least represented industries
most_represented_industry = ind_counts.idxmax()
least_represented_industry = ind_counts.idxmin()
# Display the results
print(f"The most represented industry is: {most_represented_industry} with {ind_counts.max()} companies.")
print(f"The least represented industry is: {least_represented_industry} with {ind_counts.min()} companies.")
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 8))
ind_counts.plot(kind='bar')
plt.title('Representation of Industries in the INC 5000 List')
plt.xlabel('Industry')
plt.ylabel('Number of Companies')
plt.xticks(rotation=90)
plt.show()
```

**Which industries saw the largest average growth rate?**

```python
data['Growth Rate'] = data['growth_%']
avg_growth_by_industry = data.groupby('industry')['Growth Rate'].mean().sort_values(ascending=False)
avg_growth_by_industry
```
