```python
!pip install pandas
!pip install numpy
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv('/content/financial data sp500 companies.csv')

# Display the first few rows
df.head()
```

```python
# Get the shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Get summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Data types of the columns
df.info()
```

```python
# Drop rows with missing values (or handle them accordingly)
df = df.dropna()# Histograms for numeric columns
df.hist(bins=30, figsize=(15, 10))
plt.suptitle('Distribution of Numeric Features')
plt.show()
```

```python
# Scatter plot for two numeric variables
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Net Income', y='Total Operating Expenses', hue='Gross Profit')
plt.title('Scatter Plot of Net Income vs Total Expenses')
plt.show()
```
