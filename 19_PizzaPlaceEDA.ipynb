```python
!pip install pandas
!pip install numpy
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import datetime as dt
```

```python
data_1=pd.read_csv("/content/order_details.csv")
data_1
```

```python
data_2=pd.read_csv("/content/orders.csv")
data_2
```

```python
data_3=pd.read_csv("/content/pizza_types.csv",encoding='latin')
data_3
```

```python
data_4=pd.read_csv("/content/pizzas.csv")
data_4
```

```python
cdata=data_1.merge(data_2,how='inner',on='order_id')
cdata
```

```python
cdata2=cdata.merge(data_4,how='inner',on='pizza_id')
cdata2
```

```python
data=cdata2.merge(data_3,how='inner',on='pizza_type_id')
data
```

```python
#Display the first few rows of the dataset
data.head()
```

```python
# Get basic information about the dataset
data.info()
```

```python
# Summary statistics of the dataset
data.describe()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
data.set_index(['order_details_id'])
```

```python
data['hour'] = pd.to_datetime(data['time'], infer_datetime_format=True).dt.hour
data.info()
```

```python
data['Months'] = pd.to_datetime(data['date'], infer_datetime_format = True).dt.month
data.info()
```

```python
sales_times = data['date'].value_counts()
sales_times
```

```python
sales_hours = data['hour'].value_counts().sort_values(ascending=False)
sales_hours
```

How many customers do we have each day? Are there any peak hours?



```python
total_customers=data.groupby('date')['order_id'].value_counts().sum()
total_customers
```

```python
sales_hours.head(10).to_frame().plot(kind='bar',color='purple',figsize=(10,7))
plt.xlabel('Hours')
plt.ylabel('Total_Sales')
plt.title('Peak_Time_Chart')
```

How many pizzas are typically in order? Do we have any bestsellers?



```python
average_pizzas_per_order = data.groupby('order_id')['pizza_id'].nunique().mean()
print(f"Average Pizza's per order is: {average_pizzas_per_order}")
```

```python
bestsellers = data['pizza_id'].value_counts().head(10)
plt.figure(figsize=(10,7))
bestsellers.plot(kind='bar',color='green', title='Top 10 Bestsellers')
plt.show()
```

```python
data['price'].sum()
```

```python
data['Months'].value_counts().sort_values(ascending=False)
```

```python
data.groupby('Months').agg('Months').count().to_frame().plot(kind='bar',color='yellow', figsize=(15,8))
plt.ylabel('Number of pizzas orderd per Month')
plt.title('Number of pizzas sold per Month')
```

```python
popularity=data['pizza_id'].value_counts().tail(10)
popularity
```

```python
plt.figure(figsize=(10,6))
popularity.plot(kind='bar',color='skyblue', title='Pizza Popularity')
plt.ylabel('orders')
plt.show()
```
