**  Distribution of Ages in the Dataset**

```python
# Plotting age distribution
plt.figure(figsize=(10,6))
sns.histplot(df['AGE'], bins=20, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**Smoking Status vs Lung Cancer**

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='SMOKING', hue='LUNG_CANCER')
plt.title('Smoking Status vs Lung Cancer')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='GENDER')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='GENDER', hue='LUNG_CANCER')
plt.title('Lung Cancer Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='SMOKING', hue='LUNG_CANCER')
plt.title('Lung Cancer Distribution by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='ANXIETY', hue='LUNG_CANCER')
plt.title('Lung Cancer Distribution by Anxiety Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='AGE', y='SMOKING', hue='LUNG_CANCER')
plt.title('Relationship Between Age and Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoking Status')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='ANXIETY', y='SHORTNESS OF BREATH', hue='LUNG_CANCER')
plt.title('Relationship Between Anxiety and Shortness of Breath')
plt.xlabel('Anxiety Level')
plt.ylabel('Shortness of Breath')
plt.show()
```
