```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/UltimateClassicRock.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

# Count the number of tracks per year
track_counts = df['Year'].value_counts().sort_index()

# Create the plot
plt.figure(figsize=(15, 6))
track_counts.plot(kind='bar', color='steelblue', width=0.8)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Track Count')
plt.title('Track Count by Year')

# Show the plot
plt.show()
```

```python
# Count tracks per artist
artist_track_count = df['Artist'].value_counts()

# Get the top 20 artists
top_20_artists = artist_track_count.nlargest(20)

# Plotting
plt.figure(figsize=(12, 8))
top_20_artists.plot(kind='bar', color='steelblue', figsize=(10, 6))
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.title('Top 20 Artists by Number of Tracks')

plt.show()
```

**Bob Dylan's Albums**

```python
bob_dylan_df = df[df['Artist'] == 'Bob Dylan']

# Get unique album names with their release years
bob_dylan_albums_years = bob_dylan_df[['Album', 'Year']].drop_duplicates()

# Sort by release year if needed
bob_dylan_albums_years = bob_dylan_albums_years.sort_values(by='Year')

# Print the list of album names with release years
for index, row in bob_dylan_albums_years.iterrows():
    print(f"{row['Album']} ({row['Year']})")
```

Bottom 20 Artists

```python
# Count tracks per artist
artist_track_count = df['Artist'].value_counts()

# Get the bottom 20 artists
bottom_20_artists = artist_track_count.nsmallest(20)

# Plotting
plt.figure(figsize=(12, 8))
bottom_20_artists.plot(kind='barh', color='steelblue', figsize=(10, 6))
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.title('Bottom 20 Artists by Number of Tracks')
plt.gca().invert_yaxis()  # Invert y-axis to have the artist with the fewest tracks at the top
plt.show()
```
