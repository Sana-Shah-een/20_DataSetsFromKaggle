```python
import pandas as pd
import numpy as np
police_df = pd.read_csv("/content/police.csv")
display(police_df.info())
```

```python
display(police_df.head())
```

```python
police_df.stop_date.agg(['min','max'])
```

Let's change the format of the columns where necessary.

```python
# Let's convert stop_date to a datetime type for more accurate code execution.
police_df['stop_date'] = pd.to_datetime(police_df['stop_date'])
```

Let's remove non-informative or empty columns and records

```python
# The county_name column is empty. Let's check and remove it.
display(police_df['county_name'].value_counts(dropna=False))
police_df.drop(columns=['county_name'], inplace=True)
```

```python
# The driver_age and driver_age_raw columns provide the same information.
#  driver_age_raw is the year of birth, while driver_age is the driver’s age.
#  Let's remove the year of birth column as it is unnecessary.

display(police_df[['driver_age_raw','driver_age']].head())
police_df.drop(columns=['driver_age_raw'], inplace=True)
```

```python
# The search_conducted column contains information that is not needed and may also be inaccurate.
#  We will verify this by comparing it with the drugs_related_stop column, where a True value should indicate that the search constitutes a large portion of cases.
#  Let's remove the search_conducted column.

display(police_df.groupby(['search_conducted','drugs_related_stop']).agg({'drugs_related_stop':'count'}))
police_df.drop(columns=['search_conducted'], inplace=True)
```

```python
# The is_arrested column is also non-informative, as its information is represented in the stop_outcome column as "Arrest Driver" or "Arrest Passenger."
#  We will verify this and then remove the is_arrested column.
print('---All possible outcomes of stops when is_arrested == True and their counts:---')
display(police_df.loc[(police_df['is_arrested']==True),'stop_outcome'].value_counts())
police_df.drop(columns=['is_arrested'], inplace=True)
```
